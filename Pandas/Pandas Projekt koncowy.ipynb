```python
data = pd.read_csv("..\\dane\\wind.csv")
```

Ustaw poprawione daty jako indeksy. Zwróć uwagę na typ danych, powinien to być datetime64!

```python
data["Yr"] = data["Yr"] + 1900
```

```python
data["Yr_Mo_Dy"] = pd.to_datetime({
    "year": data["Yr"],
    "month": data["Mo"],
    "day": data["Dy"]
},
format="%Y-%m-%d")
```

```python
data.drop(columns=["Yr", "Mo", "Dy"], inplace=True)
```

```python
yr_mo_dy = data["Yr_Mo_Dy"]
data.drop(labels=["Yr_Mo_Dy"], axis=1, inplace = True)
data.insert(0, "Yr_Mo_Dy", yr_mo_dy)
```

Skonstruuj pd.DataFrame() o nazwie wind_stats_daily i oblicz minimalną, maksymalną, średnią prędkość wiatru i odchylenia standardowe prędkości wiatru we wszystkich lokalizacjach dla każdego dnia osobno.

```python
wind_stats_daily = pd.DataFrame()
wind_stats_daily["min"] = np.min(data, axis=1)
wind_stats_daily["max"] = np.max(data, axis=1)
wind_stats_daily["mean"] = np.mean(data, axis=1)
wind_stats_daily["std"] = np.std(data, axis=1)
wind_stats_daily.head()
```

Zmniejsz zestaw danych do częstotliwości tygodniowej dla każdej lokalizacji, agregując punkty danych, aby reprezentowały każdy tydzień (uzyskując średnie wartości).

```python
data.reset_index(inplace=True)
```

```python
data["WeekStart"] = data.apply(lambda row: row["Yr_Mo_Dy"]- datetime.timedelta(days=row["Yr_Mo_Dy"].weekday()), axis=1)
```

```python
data["WeekEnd"] = data["WeekStart"] + pd.Timedelta(days=6)
```

```python
data["Week_str"] = data["WeekStart"].dt.strftime('%Y-%m-%d') + "/" + data["WeekEnd"].dt.strftime('%Y-%m-%d')
```

```python
data.groupby(data["Week_str"]).mean().head(4)
```

Oblicz minimalne, maksymalne, średnie prędkości wiatru i odchylenia standardowe prędkości wiatru we wszystkich lokalizacjach dla każdego tygodnia, począwszy od tygodnia 2 stycznia 1961 r., obejmującego początkowe 52 tygodnie.

```python
data.head()
```

```python
one_year_data = data[(data["Yr_Mo_Dy"].dt.year == 1961) & (data["Yr_Mo_Dy"]>="1961-01-02")]
```

```python
one_year_data.drop(columns=["WeekStart", "Week_str"], inplace=True)
```

```python
one_year_data.groupby("WeekEnd").agg(["min", "max", "mean", "std"]).head()
```
